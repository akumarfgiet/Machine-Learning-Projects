#**About Coffee Shop Sales Dataset**
**Variables** - There are 3 attributes in each case of the dataset. They are:
* TransactionID - Transction ID of orders.
* Quantity - Quantity of item purhased.
* Item - Name of item purchased.

#**Making Pivot Table to aggregate orders by TransactionID**

```python
df_pivot=df.pivot_table(index="TransactionID",columns="Item",values="Quantity",aggfunc="sum").fillna(0)
df_pivot
```

#**Exploring Transformed Data**

```python
df_pivot.info()
```

#**Converting float64 into integer**

```python
df_pivot.astype(int)
```

#**Encoding the units**

```python
def encode_units(x):
  if x<=0:
    return 0
  else:
    return 1
```

```python
df_pivot=df_pivot.applymap(encode_units)
df_pivot
```

#**Finding list of frequent items**

```python
from mlxtend.frequent_patterns import apriori, association_rules
freq_items = apriori(df_pivot,max_len=2,min_support=0.01,use_colnames=True)
freq_items
```

#**Finding Association Rule**

```python
association_rules(freq_items,metric="lift",min_threshold=1)
```
