#What is Bank Churn ?
Bank churn, also known as customer attrition, is when a customer stops using a bank's services or closes their account. It's the percentage of customers who have stopped doing business with a bank. High churn rates can negatively impact a bank's reputation and profitability.

#About dataset !
This dataset is for ABC Multistate bank with following columns:

* CustomerId, unused variable.
* Surname, unused variable.
* CreditScore, used as input.
* Geography, used as input.
* Gender, used as input.
* Age, used as input.
* Tenure, used as input.
* Balance, used as input.
* Num Of Products, used as input.
* Has Credit Card, used as input.
* Is Active Member, used as input.
* Estimated Salary, used as input.
* Churn, used as the target. 1 if the client has left the bank during some period or 0 if he/she has not.

```python
import numpy as np
import pandas as pd
df=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.drop(['CustomerId','Surname'],axis=1,inplace=True)
```

```python
df
```

```python
df=pd.get_dummies(df)
```

```python
df
```

```python
df.columns
```

```python
y=df['Churn']
x=df[['CreditScore', 'Age', 'Tenure', 'Balance', 'Num Of Products',
       'Has Credit Card', 'Is Active Member', 'Estimated Salary',
      'Geography_France', 'Geography_Germany', 'Geography_Spain',
       'Gender_Female', 'Gender_Male']]
```

```python
y.value_counts()
```

##Imbalanced Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

##Balanced Dataset

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler()
```

```python
x_ros,y_ros=ros.fit_resample(x,y)
```

```python
y_ros.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_ros,y_ros,random_state=2529)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
