#**About Dataset**
The Cancer Diagnosis Prediction dataset is a comprehensive collection of medical data aimed at predicting the likelihood of a cancer diagnosis based on various measurable features of cell nuclei present in breast tissue samples. The dataset is centered around the target variable 'diagnosis', which indicates whether the tumor is malignant or benign.

To aid in the prediction, the dataset includes multiple feature variables that describe the physical characteristics of the cell nuclei:

- **Mean Features**: These include variables such as 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', and 'fractal_dimension_mean'. These features represent the average values of the respective metrics across all cells in the sample.

- **Standard Error Features (SE)**: The variables 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', and 'fractal_dimension_se' provide insights into the variability or uncertainty in the measurements of the mean features.

- **Worst Features**: These include 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', and 'fractal_dimension_worst'. These features capture the worst (i.e., largest) values of the respective metrics found in any cell within the sample.

By analyzing these features, this dataset facilitates the development of machine learning models that can accurately predict the presence of malignant or benign tumors, contributing to early detection and treatment of breast cancer.

In the context of the cancer diagnosis dataset, the letters 'M' and 'B' in the 'diagnosis' variable represent the type of tumor:

- **'M'** stands for **Malignant**: This indicates that the tumor is cancerous and has the potential to grow aggressively and spread to other parts of the body.

- **'B'** stands for **Benign**: This indicates that the tumor is non-cancerous, generally does not spread, and is usually considered less dangerous.

These labels are crucial for distinguishing between harmful and non-harmful tumors in the dataset, allowing for the development of predictive models that can assist in medical decision-making.

#**Defining target variable(y) and feature variable(x)**

```python
df.columns
```

```python
y = df['diagnosis']
x = df[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

#**Train-Test Split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, train_size=0.8, random_state=2529 )
```

#**Model Selection**

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

#**Training the model**

```python
rfc.fit(x_train,y_train)
```

#**Testing the model**

```python
y_pred = rfc.predict(x_test)
```

#**Calculating the metrics**

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
