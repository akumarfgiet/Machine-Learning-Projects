#About Dataset
The dataset contains 1338 observations (rows) and 7 features (columns). The dataset contains 4 numerical features (age, bmi, children and id) and 3 nominal features (sex, smoker and region) that were converted into factors with numerical value designated for each level.
But here, 'id' is irrelevant variable.

The purposes of this exercise to look into different features to observe their relationship, and plot a multiple linear regression based on several features of individual such as age, physical/family condition and location against their existing medical expense to be used for predicting future medical expenses of individuals that help medical insurance to make decision on charging the premium.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Insurance%20Premium.csv")
df
```

```python
y=df['Premium']
x=df[['Age', 'Gender', 'BMI', 'Children', 'Smoker', 'Region']]
```

```python
x=pd.get_dummies(x,columns=['Gender','Region'],dtype=np.int32)
x
```

```python
x['Smoker']=x['Smoker'].replace({'yes':0,'no':1})
x
```

```python
import statsmodels.api as sm
model=sm.OLS(y,x)
print(model.fit().summary())
```

###According to above EDA model, there is no irrelevant feature variable in the model.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

###Since, r2 score = 74% means, all the feature variables are only able to explain 74% of the target variable of the model.
###Hence, there are some more feature variables required for better predictive model.
