#**Splitting bundled items into individual items**

```python
transaction=df['Transaction'].apply(lambda str:str.split(','))
transaction_list=list(transaction)
transaction_list
```

#**Encoding the transactions**

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
one_hot=te.fit(transaction).transform(transaction)
df_one_hot=pd.DataFrame(one_hot,columns=te.columns_)
df_one_hot
```

#**Finding list of frequent items**

```python
from mlxtend.frequent_patterns import apriori, association_rules
freq_items = apriori(df_one_hot,max_len=2,min_support=0.001,use_colnames=True)
freq_items
```

#**Finding Association Rule**

```python
association_rules(freq_items,metric="lift",min_threshold=0.5)
```
