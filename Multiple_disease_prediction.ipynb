#**About Dataset !**
###Prognosis prediction is a model that uses predictors available at the time of prediction to forecast a future health outcome. A prognosis is a medical assessment that predicts the course and outcome of a medical condition, or the likelihood of recovering from an injury or disease. Prognosis can vary based on age, race, sex, injury, disease, and treatment.
###This Dataset contains 'prognosis' column having 42 diseases and remaining 132 columns for 'symptoms'.

```python
import pandas as pd
df=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/MultipleDiseasePrediction.csv")
df
```

```python
y=df['prognosis']
x=df.drop(['prognosis'],axis=1)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(x_train)
x_train_new=ss.transform(x_train)
x_test_new=ss.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
```

```python
knc.fit(x_train,y_train)
```

```python
y_pred=knc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```
